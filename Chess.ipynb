# Шахматный турнир

Считываем матрицу из файла

```python
import numpy as np
import pandas as pd

home_data = pd.read_csv('data1.csv', sep=';')

n=len(home_data)


print("Турнирная таблица: ")
home_data
```

Добавляем вектор побед и сортируем

```python
vect = np.ones(n)

win_vect = np.dot(home_data,vect)
home_data['Wins'] = win_vect
home_data.sort_values(['Wins'], ascending=[0])

```

Несколько игроков имеют одинаковое количество очков, добавим коэффициент Бергера

```python
Berg_vect = np.zeros(n)


for i in range(n):
    for j in range(n):
    
        if home_data.values[i][j]==1:
            Berg_vect[i]+=home_data.values[j][n]
            
        elif home_data.values[i][j]==1/2:
            Berg_vect[i]+=(home_data.values[j][n])/2
            
home_data['Berg'] = Berg_vect

home_data.sort_values(['Wins', 'Berg'], ascending=[False, False])
```

Добавим усредненный коэффициент

```python
def berger(home_data):

    Berg_vect = np.zeros(n)
    Berg_vect_min=np.zeros(n)


    for i in range(n):
        Berg_vect_min[i]=100
        for j in range(n):
    
            if home_data.values[i][j]==1:
                Berg_vect[i]+=home_data.values[j][n]
                if Berg_vect_min[i]>home_data.values[j][n]:
                    Berg_vect_min[i]=home_data.values[j][n]
            
            elif home_data.values[i][j]==1/2:
                Berg_vect[i]+=(home_data.values[j][n])/2
                if Berg_vect_min[i]>(home_data.values[j][n])/2:
                    Berg_vect_min[i]=(home_data.values[j][n])/2
         
        
    Berg_vect=Berg_vect-Berg_vect_min
    home_data['Berg 2'] = Berg_vect
    home_data


berger(home_data)
home_data.sort_values(['Wins', 'Berg', 'Berg 2'], ascending=[False, False, False])
```

Kendall - Wei

```python
home_data_for_KW = pd.read_csv('data1.csv', sep=';')
vect = np.ones(n)
delta = 1e-8

def matrix_eigenvalue(matr, vect):
    tmp = np.dot(matr,vect)
    return np.dot(vect, tmp)

def KW(A):
    vect = np.ones(n)
    win_vect = np.dot(A,vect)
    r_1 = win_vect
    lam = matrix_eigenvalue(A, r_1)

    while True:
        Ar = np.dot(A,r_1)
        r_1_norm = Ar / np.linalg.norm(Ar)
        lam_new = matrix_eigenvalue(A, r_1_norm)
        
        error = np.abs(lam-lam_new)
        if (error < delta):
            break
            
        r_1 = r_1_norm
        lam = lam_new

    return r_1_norm

V = KW(home_data_for_KW.values)
V = V / np.dot(V,vect)
print("Собственный вектор: \n", V)
```

```python
home_data['KW'] = V
home_data.sort_values(['Wins', 'Berg', 'Berg 2'], ascending=[False, False, False])
```
